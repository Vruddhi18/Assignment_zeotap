```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load datasets

```python
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
transactions = pd.read_csv("Transactions.csv")
```

# Quick overview of data

```python
print(customers.head(), products.head(), transactions.head())
```

# Check missing values

```python
print(customers.isnull().sum(), products.isnull().sum(), transactions.isnull().sum())
```

# Convert date columns to datetime format


```python
customers["SignupDate"] = pd.to_datetime(customers["SignupDate"])
transactions["TransactionDate"] = pd.to_datetime(transactions["TransactionDate"])
```

# Merge datasets


```python
merged_df = transactions.merge(customers, on="CustomerID").merge(products, on="ProductID")
```

# EDA Visualizations


```python
plt.figure(figsize=(10, 5))
sns.countplot(data=customers, x="Region")
plt.title("Customer Distribution by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=products, x="Category", y="Price")
plt.title("Average Price of Products by Category")
plt.xticks(rotation=45)
plt.show()
```
